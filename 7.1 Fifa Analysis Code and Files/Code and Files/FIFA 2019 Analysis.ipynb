### Now let us see which clubs have the best players

Creating a DataFrame for the total ratings for all the players

```python
best_player = {}
for club in df['Club'].unique():
    overall_rating = df['Overall'][df['Club'] == club].sum()
    best_player[club] = overall_rating
df_bsp = pd.DataFrame.from_dict(best_player , orient = 'index' , 
                                                 columns = ['overall'])
df_bsp['club'] = df_bsp.index
df_bsp = df_bsp.sort_values(by = 'overall' , ascending =  False)
```

```python
df_bsp
```

```python
plt.figure(1 , figsize = (35 , 10))
sns.barplot(x = 'club' , y  = 'overall' , data = df_bsp.head(20))
plt.ylim(2350 , 2600)
plt.show()
```

### Evaluating players

Let us first list the abilities which we have as features in our dataset as a new list

```python
abilities = ['Crossing', 'Finishing', 'HeadingAccuracy',
       'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed',
       'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping',
       'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions',
       'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking',
       'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
```

Observing players on 3 most important abilities:
- Shot Power
- Long Pass
- Vision

Top players with highest shot power

```python
df.sort_values(by = 'ShotPower' , ascending = False)[['Name' , 'Club' , 'Nationality' , 
                                                     'ShotPower' ]].head(5)
```

Top players with best Long Pass

```python
df.sort_values(by = 'LongPassing' , ascending = False)[['Name' , 'Club' , 'Nationality' , 
                                                     'LongPassing']].head(5)
```

Top players with best Vision

```python
df.sort_values(by = 'Vision' , ascending = False)[['Name' , 'Club' , 'Nationality' , 
                                                     'Vision' ]].head(5)
```

Now we will find the top abilities a player should have according to the position he's playing

```python
df_postion  = pd.DataFrame()
for position_name, features in df.groupby(df['Position'])[abilities].mean().iterrows():
    top_features = dict(features.nlargest(5))
    df_postion[position_name] = tuple(top_features)
df_postion.head()
```
